## Section 1 — Variational Autoencoders (Reminder)

### 1.1  The ELBO

Our goal is to train a generative model by maximising the log-likelihood
$\log p_{\boldsymbol{\theta}}(\boldsymbol{x})$ of observed data.
The marginal

$$p(\boldsymbol{x}) = \int p_{\boldsymbol{\theta}}(\boldsymbol{x}|\boldsymbol{h})\,p(\boldsymbol{h})\,d\boldsymbol{h}$$

is intractable.  Variational inference introduces an approximate posterior
$q_{\boldsymbol{\phi}}(\boldsymbol{h}|\boldsymbol{x})$ and derives the
**Evidence Lower BOund (ELBO)**:

$$\log p(\boldsymbol{x}) \geq
\underbrace{\mathbb{E}_{q_{\boldsymbol{\phi}}(\boldsymbol{h}|\boldsymbol{x})}
\!\left[\log p_{\boldsymbol{\theta}}(\boldsymbol{x}|\boldsymbol{h})\right]}_{\text{reconstruction}}
-\underbrace{D_{\mathrm{KL}}\!\bigl(q_{\boldsymbol{\phi}}(\boldsymbol{h}|\boldsymbol{x})
\,\|\,p(\boldsymbol{h})\bigr)}_{\text{regularisation}}
=: \mathcal{L}(\boldsymbol{\phi},\boldsymbol{\theta})$$

The bound is tight when $q_{\boldsymbol{\phi}} = p(\boldsymbol{h}|\boldsymbol{x})$.

### 1.2  Reparameterisation trick

With Gaussian encoder $q_{\boldsymbol{\phi}}(\boldsymbol{h}|\boldsymbol{x})
= \mathcal{N}(\boldsymbol{\mu}_{\boldsymbol{\phi}},\boldsymbol{\sigma}^2_{\boldsymbol{\phi}}\mathbf{I})$
and prior $p(\boldsymbol{h})=\mathcal{N}(\mathbf{0},\mathbf{I})$:

$$\boldsymbol{h} = \boldsymbol{\mu}_{\boldsymbol{\phi}}(\boldsymbol{x})
  + \boldsymbol{\sigma}_{\boldsymbol{\phi}}(\boldsymbol{x})\odot\boldsymbol{\epsilon},
  \qquad \boldsymbol{\epsilon}\sim\mathcal{N}(\mathbf{0},\mathbf{I})$$

The closed-form KL divergence:

$$D_{\mathrm{KL}}(q_{\boldsymbol{\phi}}\,\|\,p)
= \tfrac{1}{2}\sum_{j=1}^{d_h}\!\left(\mu_j^2+\sigma_j^2-\log\sigma_j^2-1\right)$$

### 1.4  VAE architecture

```python
# ── VAE Model ─────────────────────────────────────────────────────────────
class VAE(nn.Module):
    """
    Convolutional VAE for 28x28 grayscale images (MNIST).

    Encoder:  x (1x28x28) -> Conv stack -> mu, log_var  (LATENT_DIM,)
    Decoder:  z (LATENT_DIM,) -> ConvTranspose stack -> x_hat (1x28x28)

    The decoder outputs logits for Bernoulli(sigmoid(logit)), so we use
    binary cross-entropy reconstruction loss.
    """
    def __init__(self, latent_dim: int = 20):
        super().__init__()
        self.latent_dim = latent_dim

        # ── Encoder ────────────────────────────────────────────────────────
        # Input: (B, 1, 28, 28)
        self.encoder = nn.Sequential(
            nn.Conv2d(1, 32, 4, stride=2, padding=1),   # -> (B,32,14,14)
            nn.LeakyReLU(0.2),
            nn.Conv2d(32, 64, 4, stride=2, padding=1),  # -> (B,64, 7, 7)
            nn.LeakyReLU(0.2),
            nn.Flatten(),                                # -> (B, 64*7*7)
        )
        enc_flat = 64 * 7 * 7
        self.fc_mu      = nn.Linear(enc_flat, latent_dim)
        self.fc_log_var = nn.Linear(enc_flat, latent_dim)

        # ── Decoder ────────────────────────────────────────────────────────
        self.fc_dec = nn.Linear(latent_dim, enc_flat)
        self.decoder = nn.Sequential(
            nn.Unflatten(1, (64, 7, 7)),
            nn.ConvTranspose2d(64, 32, 4, stride=2, padding=1),  # -> (B,32,14,14)
            nn.LeakyReLU(0.2),
            nn.ConvTranspose2d(32,  1, 4, stride=2, padding=1),  # -> (B,1,28,28)
            # No final activation — we output logits for BCEWithLogitsLoss
        )

    # -- Inference network ------------------------------------------------
    def encode(self, x):
        """x: (B,1,28,28) -> mu, log_var: (B, latent_dim)"""
        h = self.encoder(x)
        return self.fc_mu(h), self.fc_log_var(h)

    # -- Reparameterisation trick -----------------------------------------
    def reparameterise(self, mu, log_var):
        """
        h = mu + sigma * eps,  eps ~ N(0,I)
        std = exp(0.5 * log_var)  ->  var = sigma^2
        """
        std = torch.exp(0.5 * log_var)
        eps = torch.randn_like(std)
        return mu + std * eps

    # -- Generative network -----------------------------------------------
    def decode(self, z):
        """z: (B, latent_dim) -> logits: (B,1,28,28)"""
        h = self.fc_dec(z)
        return self.decoder(h)

    def forward(self, x):
        mu, log_var = self.encode(x)
        z           = self.reparameterise(mu, log_var)
        x_hat       = self.decode(z)
        return x_hat, mu, log_var


# ── ELBO loss ──────────────────────────────────────────────────────────────
def vae_loss(x, x_hat_logits, mu, log_var, beta: float = 1.0):
    """
    ELBO = -Reconstruction + beta * KL

    Reconstruction:  BCE(sigmoid(x_hat_logits), x)   [pixels in [0,1]]
    KL:              0.5 * sum(mu^2 + sigma^2 - log_sigma^2 - 1)

    beta controls the KL weight (beta-VAE; beta=1 is the standard VAE).
    """
    B = x.size(0)
    # Flatten for BCE
    recon = F.binary_cross_entropy_with_logits(
        x_hat_logits.view(B, -1), x.view(B, -1), reduction='sum') / B
    # Closed-form KL: -0.5 * sum(1 + log_var - mu^2 - exp(log_var))
    kl = -0.5 * (1 + log_var - mu.pow(2) - log_var.exp()).sum() / B
    return recon + beta * kl, recon, kl


# Instantiate
vae = VAE(latent_dim=20).to(device)
n_params_vae = sum(p.numel() for p in vae.parameters() if p.requires_grad)
print(f'VAE parameters: {n_params_vae:,}')
print(vae)
```

### 1.5  VAE training

```python
# ── Enable cuDNN autotuner (free speed for fixed input sizes) ─────────────
torch.backends.cudnn.benchmark = True

def train_vae(model, loader, epochs=20, lr=1e-3, beta=1.0):
    """
    Train VAE.  Key efficiency improvements over the baseline:

    1. zero_grad(set_to_none=True)  — deallocates gradient memory instead of
       zeroing it; saves one memset per parameter tensor per step.
    2. AMP autocast (GPU only)      — fp16 compute with fp32 master weights;
       roughly 2x throughput on modern Ampere/Ada GPUs.
    3. Cosine LR schedule           — often converges in fewer epochs.
    4. scaler.scale / unscale       — stable fp16 gradient scaling.
    """
    opt       = torch.optim.Adam(model.parameters(), lr=lr)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(opt, T_max=epochs)
    history   = {'elbo': [], 'recon': [], 'kl': []}
    t0        = time.time()

    for ep in range(1, epochs + 1):
        model.train()
        tot_elbo = tot_recon = tot_kl = 0.0

        for x, _ in loader:
            x = x.to(device, non_blocking=True)   # non_blocking overlaps H2D copy

            with autocast:
                x_hat, mu, lv = model(x)
                loss, recon, kl = vae_loss(x, x_hat, mu, lv, beta=beta)

            scaler.scale(loss).backward()
            scaler.step(opt)
            scaler.update()
            opt.zero_grad(set_to_none=True)        # ← key: avoids memset

            tot_elbo  += loss.item()
            tot_recon += recon.item()
            tot_kl    += kl.item()

        scheduler.step()
        n = len(loader)
        history['elbo'].append(tot_elbo / n)
        history['recon'].append(tot_recon / n)
        history['kl'].append(tot_kl / n)

        if ep % 5 == 0 or ep == 1:
            print(f'  Epoch {ep:3d}/{epochs}  '
                  f'ELBO={history["elbo"][-1]:.2f}  '
                  f'Recon={history["recon"][-1]:.2f}  '
                  f'KL={history["kl"][-1]:.2f}  '
                  f'lr={scheduler.get_last_lr()[0]:.2e}')

    elapsed = time.time() - t0
    print(f'Training time: {elapsed:.1f}s  '          f'({elapsed/epochs:.2f}s/epoch)')
    history['wall_s'] = elapsed
    return history


# ── Optional: torch.compile (PyTorch >= 2.0, GPU only) ────────────────────
# compile() traces the model once and emits optimised CUDA kernels.
# Use only if you are not actively modifying the model.
if device_str == 'cuda' and hasattr(torch, 'compile'):
    vae = torch.compile(vae)
    print('VAE compiled with torch.compile()')

print('Training VAE on MNIST (20 epochs)...')
vae_history = train_vae(vae, loader_train_01, epochs=20, lr=1e-3, beta=1.0)
```

## Section 2 — Denoising Diffusion Probabilistic Models (DDPM)

### 2.1  Mathematical foundations

The DDPM (Ho et al. 2020) defines:

**Forward process** — a fixed Markov chain that gradually adds Gaussian noise:
$$q(\boldsymbol{x}_t|\boldsymbol{x}_{t-1})
= \mathcal{N}(\boldsymbol{x}_t;\,\sqrt{\alpha_t}\,\boldsymbol{x}_{t-1},\,(1-\alpha_t)\mathbf{I})$$

**Closed-form marginal** (key derivation — see slides):
$$\boxed{q(\boldsymbol{x}_t|\boldsymbol{x}_0)
= \mathcal{N}(\boldsymbol{x}_t;\,\sqrt{\bar\alpha_t}\,\boldsymbol{x}_0,\,
(1-\bar\alpha_t)\mathbf{I})}
\qquad \bar\alpha_t = \prod_{s=1}^t\alpha_s$$

**Training objective** — predict the noise $\boldsymbol{\epsilon}$ that was added:
$$\mathcal{L}_{\text{simple}}
= \mathbb{E}_{t,\boldsymbol{x}_0,\boldsymbol{\epsilon}}\!\left[
  \|\boldsymbol{\epsilon}
    - \boldsymbol{\epsilon}_{\boldsymbol{\theta}}(\sqrt{\bar\alpha_t}\boldsymbol{x}_0
      + \sqrt{1-\bar\alpha_t}\boldsymbol{\epsilon},\,t)\|^2\right]$$

**Reverse step** — DDPM sampler:
$$\boldsymbol{x}_{t-1} =
\frac{1}{\sqrt{\alpha_t}}\!\left(\boldsymbol{x}_t
- \frac{\beta_t}{\sqrt{1-\bar\alpha_t}}\,\boldsymbol{\epsilon}_{\boldsymbol{\theta}}(\boldsymbol{x}_t,t)\right)
+ \sqrt{\tilde\beta_t}\,\boldsymbol{z},\quad
\boldsymbol{z}\sim\mathcal{N}(\mathbf{0},\mathbf{I})$$


### 2.2  Noise schedule and forward process

```python
# ── Diffusion hyperparameters ──────────────────────────────────────────────
T          = 300
BETA_START = 1e-4
BETA_END   = 0.02
IMG_SIZE   = 28
CHANNELS   = 1

# ── Noise schedule — all tensors on device from the start ─────────────────
# Storing on device eliminates a host→device copy inside every training step.
betas               = torch.linspace(BETA_START, BETA_END, T, device=device)
alphas              = 1.0 - betas
alphas_cumprod      = torch.cumprod(alphas, dim=0)
alphas_cumprod_prev = F.pad(alphas_cumprod[:-1], (1, 0), value=1.0)

sqrt_acp        = alphas_cumprod.sqrt()
sqrt_1m_acp     = (1.0 - alphas_cumprod).sqrt()
sqrt_recip_a    = (1.0 / alphas).sqrt()
posterior_var   = betas * (1.0 - alphas_cumprod_prev) / (1.0 - alphas_cumprod)

print(f'T={T}, beta in [{BETA_START:.1e}, {BETA_END:.2f}]')
print(f'  sqrt(ᾱ) at t=0   : {float(sqrt_acp[0]):.4f}')
print(f'  sqrt(ᾱ) at t=T/2 : {float(sqrt_acp[T//2]):.4f}')
print(f'  sqrt(ᾱ) at t=T-1 : {float(sqrt_acp[-1]):.4f}')

# ── Forward process q(x_t | x_0) ─────────────────────────────────────────
def q_sample(x0, t, noise=None):
    """
    Sample x_t = sqrt(ᾱ_t)·x_0 + sqrt(1-ᾱ_t)·ε  in one vectorised op.
    t: (B,) integer tensor already on device.
    """
    if noise is None:
        noise = torch.randn_like(x0)
    # Direct tensor indexing — no Python loop, no torch.tensor() overhead
    s_acp = sqrt_acp[t].view(-1, 1, 1, 1)
    s_1m  = sqrt_1m_acp[t].view(-1, 1, 1, 1)
    return s_acp * x0 + s_1m * noise, noise


x0_demo = next(iter(loader_train_11))[0][:1].to(device)
timesteps_show = [0, 49, 99, 149, 199, 249, 299]
noised_imgs = []
for t_val in timesteps_show:
    t_tensor = torch.tensor([t_val], device=device)
    xt, _    = q_sample(x0_demo, t_tensor)
    noised_imgs.append(xt.squeeze().cpu())

fig, axes = plt.subplots(1, len(timesteps_show), figsize=(14, 2.5))
for ax, img, t_val in zip(axes, noised_imgs, timesteps_show):
    ax.imshow(img.clamp(-1,1)*0.5+0.5, cmap='gray', vmin=0, vmax=1)
    ax.set_title(f't={t_val}', fontsize=11); ax.axis('off')
fig.suptitle(r'Forward diffusion: $q(\mathbf{x}_t|\mathbf{x}_0)$',
             fontsize=13, fontweight='bold')
plt.tight_layout(); plt.show()
```

### 2.3  Sinusoidal time embedding

```python
class SinusoidalPosEmb(nn.Module):
    """
    Sinusoidal time embedding — frequencies registered as a buffer so they
    are computed once at construction and never recomputed during forward.
    """
    def __init__(self, dim: int):
        super().__init__()
        half  = dim // 2
        freqs = torch.exp(
            -math.log(10000) *
            torch.arange(half, dtype=torch.float32) / max(half - 1, 1)
        )
        # register_buffer ensures freqs moves with model.to(device)
        if hasattr(self, 'register_buffer'):
            self.register_buffer('freqs', freqs)
        else:
            self.freqs = freqs   # fallback for environments without register_buffer

    def forward(self, t):            # t: (B,) float in [0, 1]
        args = t.unsqueeze(1) * self.freqs.unsqueeze(0)   # (B, half)
        return torch.cat([args.sin(), args.cos()], dim=-1) # (B, dim)


emb_test = SinusoidalPosEmb(128)(torch.linspace(0, 1, 10))
print(f'Sinusoidal embedding: input (10,) → output {emb_test.shape}')
```

### 2.4  U-Net denoiser architecture

```python
class ResBlock(nn.Module):
    """
    Residual conv block with GroupNorm and time-embedding injection.
    Implements: x -> Conv -> GN -> SiLU -> Conv -> GN + time_proj + skip
    """
    def __init__(self, in_ch, out_ch, emb_dim):
        super().__init__()
        self.conv1 = nn.Sequential(
            nn.Conv2d(in_ch, out_ch, 3, padding=1),
            nn.GroupNorm(8, out_ch), nn.SiLU()
        )
        self.time_proj = nn.Sequential(nn.SiLU(), nn.Linear(emb_dim, out_ch))
        self.conv2 = nn.Sequential(
            nn.Conv2d(out_ch, out_ch, 3, padding=1),
            nn.GroupNorm(8, out_ch), nn.SiLU()
        )
        self.res_conv = nn.Conv2d(in_ch, out_ch, 1) if in_ch != out_ch else nn.Identity()

    def forward(self, x, t_emb):
        h = self.conv1(x)
        h = h + self.time_proj(t_emb)[:, :, None, None]
        h = self.conv2(h)
        return h + self.res_conv(x)


class UNet(nn.Module):
    """
    Lightweight U-Net for noise prediction on 28x28 images.

    Architecture:
        Encoder: 3 ResBlocks with downsampling
        Bottleneck: 1 ResBlock
        Decoder: 3 ResBlocks with upsampling + skip connections
        Output conv: -> (C, H, W) predicted noise

    Time conditioning: sinusoidal embedding -> MLP -> injected in every ResBlock.
    """
    def __init__(self, in_ch=1, base_ch=64, emb_dim=128):
        super().__init__()
        self.time_mlp = nn.Sequential(
            SinusoidalPosEmb(emb_dim),
            nn.Linear(emb_dim, emb_dim * 2),
            nn.SiLU(),
            nn.Linear(emb_dim * 2, emb_dim),
        )
        ch = base_ch
        # ── Encoder ────────────────────────────────────────────────────────
        self.enc1 = ResBlock(in_ch, ch,      emb_dim)   # 28x28
        self.down1 = nn.Conv2d(ch, ch,  3, stride=2, padding=1)  # 14x14
        self.enc2 = ResBlock(ch,  ch*2,  emb_dim)   # 14x14
        self.down2 = nn.Conv2d(ch*2, ch*2, 3, stride=2, padding=1) # 7x7
        self.enc3 = ResBlock(ch*2, ch*4, emb_dim)   # 7x7
        # ── Bottleneck ─────────────────────────────────────────────────────
        self.mid  = ResBlock(ch*4, ch*4, emb_dim)   # 7x7
        # ── Decoder ────────────────────────────────────────────────────────
        self.up2  = nn.ConvTranspose2d(ch*4, ch*2, 4, stride=2, padding=1)  # 14x14
        self.dec2 = ResBlock(ch*4, ch*2, emb_dim)   # 14x14 (with skip)
        self.up1  = nn.ConvTranspose2d(ch*2, ch,   4, stride=2, padding=1)  # 28x28
        self.dec1 = ResBlock(ch*2, ch,   emb_dim)   # 28x28 (with skip)
        self.out  = nn.Conv2d(ch, in_ch, 1)

    def forward(self, x, t):
        """x: (B,C,H,W),  t: (B,) float in [0,1]"""
        te = self.time_mlp(t)          # (B, emb_dim)
        # Encoder
        e1 = self.enc1(x,  te)                        # (B, ch, 28, 28)
        e2 = self.enc2(self.down1(e1), te)             # (B,2ch, 14, 14)
        e3 = self.enc3(self.down2(e2), te)             # (B,4ch,  7,  7)
        # Bottleneck
        m  = self.mid(e3, te)                          # (B,4ch,  7,  7)
        # Decoder with skip connections
        d2 = self.dec2(torch.cat([self.up2(m), e2], dim=1), te)  # (B,2ch,14,14)
        d1 = self.dec1(torch.cat([self.up1(d2), e1], dim=1), te) # (B, ch,28,28)
        return self.out(d1)                            # (B,C,28,28)


ddpm_model   = UNet(in_ch=CHANNELS, base_ch=64, emb_dim=128).to(device)
n_params_ddpm = sum(p.numel() for p in ddpm_model.parameters() if p.requires_grad)
print(f'DDPM U-Net parameters: {n_params_ddpm:,}')
```

### 2.5  DDPM training

```python
def diffusion_loss(model, x0):
    """
    L_simple: sample t ~ U[0,T-1], add noise, predict noise, MSE.
    t and x_t stay on device throughout — no unnecessary transfers.
    """
    B      = x0.size(0)
    t      = torch.randint(0, T, (B,), device=device)
    x_t, noise = q_sample(x0, t)
    pred   = model(x_t, t.float() / T)
    return F.mse_loss(pred, noise)


def train_ddpm(model, loader, epochs=20, lr=2e-4):
    """
    Train DDPM with the same efficiency stack as train_vae:
    AMP · set_to_none · non_blocking · gradient clipping.
    """
    opt       = torch.optim.Adam(model.parameters(), lr=lr)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(opt, T_max=epochs)
    history   = {'loss': []}
    t0        = time.time()

    for ep in range(1, epochs + 1):
        model.train()
        tot = 0.0
        for x, _ in loader:
            x = x.to(device, non_blocking=True)

            with autocast:
                loss = diffusion_loss(model, x)

            scaler.scale(loss).backward()
            # Gradient clipping: unscale first so clip operates on true grads
            scaler.unscale_(opt)
            nn.utils.clip_grad_norm_(model.parameters(), 1.0)
            scaler.step(opt)
            scaler.update()
            opt.zero_grad(set_to_none=True)

            tot += loss.item()

        scheduler.step()
        history['loss'].append(tot / len(loader))
        if ep % 5 == 0 or ep == 1:
            print(f'  Epoch {ep:3d}/{epochs}  '                  f'MSE={history["loss"][-1]:.5f}  '                  f'lr={scheduler.get_last_lr()[0]:.2e}')

    history['wall_s'] = time.time() - t0
    print(f'Training time: {history["wall_s"]:.1f}s  '          f'({history["wall_s"]/epochs:.2f}s/epoch)')
    return history


# ── Optional: torch.compile (PyTorch >= 2.0, GPU only) ────────────────────
if device_str == 'cuda' and hasattr(torch, 'compile'):
    ddpm_model = torch.compile(ddpm_model)
    print('DDPM compiled with torch.compile()')

print('Training DDPM on MNIST (20 epochs)...')
ddpm_history = train_ddpm(ddpm_model, loader_train_11, epochs=20, lr=2e-4)
```

### 2.6  DDPM sampling

```python
@torch.no_grad()
def ddpm_sample(model, n_samples=16, verbose=True):
    """
    DDPM reverse process.

    Key efficiency fix: all schedule tensors are already on `device`,
    so slicing them (betas[i], sqrt_recip_a[i], …) returns device tensors
    without any host↔device transfer per step.
    """
    model.eval()
    x = torch.randn(n_samples, CHANNELS, IMG_SIZE, IMG_SIZE, device=device)

    for i in reversed(range(T)):
        t_in      = torch.full((n_samples,), i / T, device=device)  # scalar broadcast
        eps_hat   = model(x, t_in)

        # Direct device-tensor slices — no torch.tensor() wrapper needed
        beta_t    = betas[i]
        s_recip_a = sqrt_recip_a[i]
        s_1m_acp  = sqrt_1m_acp[i]
        pvar      = posterior_var[i]

        mu = s_recip_a * (x - beta_t / s_1m_acp * eps_hat)
        if i > 0:
            mu = mu + pvar.sqrt() * torch.randn_like(x)
        x = mu

        if verbose and i % 50 == 0:
            print(f'  t={i:4d}  x ∈ [{x.min():.2f}, {x.max():.2f}]')

    return x.clamp(-1, 1)


print('Generating 16 DDPM samples...')
ddpm_samples    = ddpm_sample(ddpm_model, n_samples=16, verbose=True)
ddpm_samples_01 = (ddpm_samples * 0.5 + 0.5).cpu()
show_images(ddpm_samples_01, 'DDPM — samples')
```

### 2.7  Visualising the reverse diffusion trajectory

```python
@torch.no_grad()
def ddpm_sample_trajectory(model, n_img=4):
    """Reverse diffusion with snapshots — schedule sliced on device."""
    model.eval()
    x       = torch.randn(n_img, CHANNELS, IMG_SIZE, IMG_SIZE, device=device)
    traj    = {}
    snap_at = {T-1, T//4*3, T//2, T//4, 10, 0}

    for i in reversed(range(T)):
        t_in    = torch.full((n_img,), i / T, device=device)
        eps_hat = model(x, t_in)

        beta_t = betas[i]
        mu     = sqrt_recip_a[i] * (x - beta_t / sqrt_1m_acp[i] * eps_hat)
        if i > 0:
            mu = mu + posterior_var[i].sqrt() * torch.randn_like(x)
        x = mu
        if i in snap_at:
            traj[i] = (x.clamp(-1, 1) * 0.5 + 0.5).cpu()

    return traj


print('Recording reverse diffusion trajectory...')
traj      = ddpm_sample_trajectory(ddpm_model, n_img=4)
snap_keys = sorted(traj.keys(), reverse=True)

fig, axes = plt.subplots(4, len(snap_keys), figsize=(14, 7))
for col, t_key in enumerate(snap_keys):
    imgs = traj[t_key]
    for row in range(4):
        axes[row, col].imshow(imgs[row].squeeze(), cmap='gray', vmin=0, vmax=1)
        axes[row, col].axis('off')
    axes[0, col].set_title(f't={t_key}', fontsize=10)

fig.suptitle(r'DDPM reverse: $x_T \to x_0$', fontsize=12, fontweight='bold')
plt.tight_layout(); plt.show()
```

## Section 3 — Head-to-Head Comparison: VAE vs DDPM

### 3.1  Mathematical parallel

Both models optimise a **variational lower bound** on $\log p(\boldsymbol{x})$:

| Aspect | VAE | DDPM |
|:-------|:----|:-----|
| Latent dimension | $d_h \ll d_x$ (bottleneck) | $d_h = d_x$ (full) |
| Number of latent levels | 1 | $T = 300$–$1000$ |
| Encoder $q$ | **Learned** $q_{\boldsymbol{\phi}}(\boldsymbol{h}|\boldsymbol{x})$ | **Fixed** Gaussian schedule |
| Reparameterisation | $\boldsymbol{h}=\boldsymbol{\mu}+\boldsymbol{\sigma}\odot\boldsymbol{\epsilon}$ | $\boldsymbol{x}_t=\sqrt{\bar\alpha_t}\boldsymbol{x}_0+\sqrt{1-\bar\alpha_t}\boldsymbol{\epsilon}$ |
| Prior | $\mathcal{N}(\boldsymbol{0},\mathbf{I})$ | $\mathcal{N}(\boldsymbol{0},\mathbf{I})$ |
| Training loss | $-\text{ELBO}=\text{BCE}+\beta\,D_{\mathrm{KL}}$ | $\mathbb{E}\|\boldsymbol{\epsilon}-\boldsymbol{\epsilon}_{\boldsymbol{\theta}}(\boldsymbol{x}_t,t)\|^2$ |
| Sampling | 1 decoder call | $T$ denoiser calls |
| Sample quality | Blurry (Gaussian averaging) | Sharp (state of the art) |
| Structured latent | ✓ (interpolation, clustering) | ✗ |
| Training stability | ✓ | ✓ |
| Mode collapse | Rare | Not observed |

**Key insight:** DDPM can be derived from the hierarchical VAE ELBO
by taking $T\to\infty$ latent levels with a fixed Gaussian encoder —
see the lecture slides for the full derivation.


### 3.2  Quantitative comparison

```python
# ── Quantitative comparison helpers ──────────────────────────────────────
def pixel_stats_fast(tensor_01, n_max=2000):
    """
    Per-pixel mean and variance (diagonal of covariance) — O(N·D) not O(D²).
    We only need the diagonal for the proxy-FID, so skip the full covariance.
    """
    imgs = tensor_01[:n_max].view(min(len(tensor_01), n_max), -1).float()
    mu  = imgs.mean(0)
    var = ((imgs - mu.unsqueeze(0))**2).mean(0)   # unbiased=False, matches real torch
    return mu.numpy(), var.numpy()

def pixel_stats_from_loader(loader, n_max=2000):
    chunks = []
    for x, _ in loader:
        chunks.append(x.view(x.size(0), -1))
        if sum(c.size(0) for c in chunks) >= n_max:
            break
    imgs = torch.cat(chunks)[:n_max].float()
    mu  = imgs.mean(0)
    var = ((imgs - mu.unsqueeze(0))**2).mean(0)   # unbiased=False, matches real torch
    return mu.numpy(), var.numpy()

def proxy_fid(mu_r, var_r, mu_g, var_g):
    """Fréchet distance using diagonal (per-pixel) covariance — O(D)."""
    diff    = float(np.sum((mu_r - mu_g)**2))
    tr_term = float(np.sum(var_r + var_g - 2.0*np.sqrt(np.maximum(var_r*var_g, 0))))
    return diff + tr_term

# ── Real test images ───────────────────────────────────────────────────────
mu_real, var_real = pixel_stats_from_loader(loader_test_01, n_max=2000)

# ── VAE: generate 2000 samples in a single forward pass ───────────────────
vae.eval()
with torch.no_grad():
    z_big   = torch.randn(2000, vae.latent_dim, device=device)
    vae_big = torch.sigmoid(vae.decode(z_big)).cpu()
mu_vae, var_vae = pixel_stats_fast(vae_big, n_max=2000)

# ── DDPM: generate 256 samples in ONE vectorised call (not a Python loop) ─
print('Generating 256 DDPM samples (one vectorised call)...')
ddpm_model.eval()
with torch.no_grad():
    ddpm_big_gpu = ddpm_sample(ddpm_model, n_samples=256, verbose=False)
ddpm_big = (ddpm_big_gpu * 0.5 + 0.5).cpu()
mu_ddpm, var_ddpm = pixel_stats_fast(ddpm_big, n_max=256)

fid_vae  = proxy_fid(mu_real, var_real, mu_vae,  var_vae)
fid_ddpm = proxy_fid(mu_real, var_real, mu_ddpm, var_ddpm)

# ── Sharpness: Laplacian variance ─────────────────────────────────────────
def laplacian_sharpness(t01):
    """
    Sharpness = variance of Laplacian response.
    Implemented with scipy.ndimage for robustness; same result as F.conv2d.
    """
    from scipy.ndimage import laplace
    imgs_np = t01[:256].squeeze(1).numpy()   # (N, H, W)
    laps    = np.stack([laplace(img) for img in imgs_np])
    return float(laps.var())

sharpness_real = laplacian_sharpness(next(iter(loader_test_01))[0][:256])
sharpness_vae  = laplacian_sharpness(vae_big[:256])
sharpness_ddpm = laplacian_sharpness(ddpm_big[:256])

hdr = f'{"Model":<18}  {"Proxy-FID":>10}  {"Sharpness":>10}  {"Train":>10}  {"Sample":>12}'
sep = '-' * len(hdr)
print(); print(hdr); print(sep)
print(f'{"Real":<18}  {"—":>10}  {sharpness_real:>10.4f}  {"—":>10}  {"—":>12}')
print(f'{"VAE":<18}  {fid_vae:>10.2f}  {sharpness_vae:>10.4f}  '      f'{vae_history["wall_s"]:>9.1f}s  {"~0.01s/img":>12}')
print(f'{"DDPM":<18}  {fid_ddpm:>10.2f}  {sharpness_ddpm:>10.4f}  '      f'{ddpm_history["wall_s"]:>9.1f}s  {f"~{T*0.002:.1f}s/img":>12}')
```

### 3.4  Critical discussion: pros, cons, and when to use each

#### Why VAE samples are blurry

The VAE decoder maximises the expected log-likelihood
$\mathbb{E}_{q_{\boldsymbol{\phi}}}[\log p_{\boldsymbol{\theta}}(\boldsymbol{x}|\boldsymbol{h})]$.
With a Gaussian (or Bernoulli) decoder, this is equivalent to minimising a pixel-wise
$\ell_2$ (or cross-entropy) loss *averaged over the posterior distribution*.
Because the posterior has nonzero variance, the decoder learns to output the
**mean over likely completions** — which is blurry when there is genuine ambiguity.
This is a fundamental limitation of maximum-likelihood training with a unimodal decoder,
not a pathology that can be fixed by deeper networks alone.

#### Why DDPM samples are sharper

DDPM never explicitly predicts the mean of $\boldsymbol{x}_0$ in a single shot.
Instead, the network predicts only the noise $\boldsymbol{\epsilon}$ at each small step.
The cumulative effect of many small, iteratively-applied denoising corrections
allows the model to commit to sharp, detailed features without averaging.
Mathematically, DDPM minimises a sum of $T$ KL terms between the true forward
posterior $q(\boldsymbol{x}_{t-1}|\boldsymbol{x}_t,\boldsymbol{x}_0)$ and the learned
reverse $p_{\boldsymbol{\theta}}(\boldsymbol{x}_{t-1}|\boldsymbol{x}_t)$ — a much richer objective than
the single-step VAE ELBO.

#### Posterior collapse in VAEs

If the decoder $p_{\boldsymbol{\theta}}(\boldsymbol{x}|\boldsymbol{h})$ is sufficiently powerful,
it can reconstruct $\boldsymbol{x}$ without using $\boldsymbol{h}$ at all, making
$q_{\boldsymbol{\phi}}(\boldsymbol{h}|\boldsymbol{x}) \approx p(\boldsymbol{h})$.
The KL term then drops to zero, the encoder collapses, and the latent space
becomes useless. DDPM has no analogous problem: the fixed forward process
guarantees that the reverse network must use the conditioning information at each step.

#### The ELBO gap

VAE training never exactly maximises $\log p(\boldsymbol{x})$ because
$\log p(\boldsymbol{x}) = \text{ELBO} + D_{\mathrm{KL}}(q_{\boldsymbol{\phi}}\|p(\boldsymbol{h}|\boldsymbol{x}))$,
and the second term is positive unless the mean-field Gaussian exactly matches
the (generally non-Gaussian) true posterior. DDPM's sum-of-KL ELBO converges
to the true $\log p(\boldsymbol{x})$ in the limit $T\to\infty$ with a fine enough schedule.

#### When to choose each model

| Use case | Recommended model | Reason |
|:---------|:------------------|:-------|
| Image / audio / video generation | **DDPM** | State-of-the-art sample quality |
| Anomaly detection | **VAE** | Likelihood evaluation is fast; latent score is meaningful |
| Representation learning / clustering | **VAE** | Structured, low-dim latent space |
| Scientific data (molecules, spectra) | **VAE** | Compact latent enables property prediction |
| Text-guided image synthesis | **LDM** (VAE + DDPM) | Combines VAE compression with diffusion quality |
| Real-time generation | **VAE** | Single decoder call vs. $T$ steps |
| Distribution coverage / diversity | **DDPM** | No mode collapse; covers full support |


### 3.5  Bonus: DDIM faster sampling

```python
@torch.no_grad()
def ddim_sample(model, n_samples=16, n_steps=50, eta=0.0):
    """
    DDIM deterministic ODE sampler — schedule tensors sliced on device.
    """
    model.eval()
    x = torch.randn(n_samples, CHANNELS, IMG_SIZE, IMG_SIZE, device=device)

    step_indices = torch.linspace(T - 1, 0, n_steps, device=device).long()

    for idx in range(len(step_indices) - 1):
        t_curr = int(step_indices[idx].item())
        t_prev = int(step_indices[idx + 1].item())

        t_in    = torch.full((n_samples,), t_curr / T, device=device)
        eps_hat = model(x, t_in)

        # Direct on-device slices
        acp_t    = alphas_cumprod[t_curr]
        acp_prev = alphas_cumprod[t_prev]

        x0_hat = (x - (1 - acp_t).sqrt() * eps_hat) / acp_t.sqrt()
        x0_hat = x0_hat.clamp(-1, 1)
        dir_xt  = (1 - acp_prev).sqrt() * eps_hat
        sigma_t = eta * ((1 - acp_prev) / (1 - acp_t) * (1 - acp_t / acp_prev)).sqrt()
        noise   = torch.randn_like(x) if (eta > 0 and t_prev > 0) else torch.zeros_like(x)
        x       = acp_prev.sqrt() * x0_hat + dir_xt + sigma_t * noise

    return x.clamp(-1, 1)


import time as _time
t0     = _time.time()
ddpm_s = ddpm_sample(ddpm_model, n_samples=8, verbose=False)
t_ddpm = _time.time() - t0
t0     = _time.time()
ddim_s = ddim_sample(ddpm_model, n_samples=8, n_steps=50, eta=0.0)
t_ddim = _time.time() - t0
print(f'DDPM ({T} steps): {t_ddpm:.2f}s  |  DDIM (50 steps): {t_ddim:.2f}s  '
      f'(~{t_ddpm/t_ddim:.0f}x speedup)')

fig, axes = plt.subplots(2, 8, figsize=(16, 5))
for col in range(8):
    for row, imgs in enumerate([(ddpm_s*0.5+0.5).cpu(), (ddim_s*0.5+0.5).cpu()]):
        axes[row, col].imshow(imgs[col].squeeze().clamp(0,1), cmap='gray')
        axes[row, col].axis('off')
axes[0, 0].set_ylabel(f'DDPM T={T}',    fontsize=10)
axes[1, 0].set_ylabel('DDIM n=50',     fontsize=10)
plt.suptitle(f'DDPM vs DDIM — same model', fontsize=12, fontweight='bold')
plt.tight_layout(); plt.show()
```

```python
# ── Exercise C1: Cosine schedule vs. linear schedule ──────────────────────
def cosine_schedule(T, s=0.008):
    """
    Nichol & Dhariwal (2021) cosine schedule.
    Avoids too-small SNR at the start of the linear schedule.
    """
    steps = T + 1
    t     = torch.linspace(0, T, steps)
    f     = torch.cos(((t / T + s) / (1 + s)) * math.pi / 2).pow(2)
    acp   = f / f[0]
    betas_cos = 1 - acp[1:] / acp[:-1]
    return betas_cos.clamp(min=1e-5, max=0.999)


betas_lin = torch.linspace(BETA_START, BETA_END, T)
betas_cos = cosine_schedule(T)
acp_lin   = torch.cumprod(1 - betas_lin, 0)
acp_cos   = torch.cumprod(1 - betas_cos, 0)

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
t_range   = range(T)

axes[0].plot(t_range, betas_lin.numpy(), label='Linear', lw=2)
axes[0].plot(t_range, betas_cos.numpy(), label='Cosine', lw=2)
axes[0].set_xlabel('t'); axes[0].set_ylabel(r'$\beta_t$')
axes[0].set_title('Noise schedules $\\beta_t$')
axes[0].legend(); axes[0].grid(True, alpha=0.3)

axes[1].plot(t_range, (acp_lin/(1-acp_lin+1e-8)).log().numpy(), label='Linear', lw=2)
axes[1].plot(t_range, (acp_cos/(1-acp_cos+1e-8)).log().numpy(), label='Cosine', lw=2)
axes[1].set_xlabel('t'); axes[1].set_ylabel(r'$\log$ SNR')
axes[1].set_title('Signal-to-Noise Ratio (log scale)')
axes[1].legend(); axes[1].grid(True, alpha=0.3)

plt.suptitle('Exercise C1: Compare linear vs cosine noise schedules',
             fontsize=12, fontweight='bold')
plt.tight_layout(); plt.show()
print('Try training a DDPM with the cosine schedule and compare sample quality.')
```

### 3.6  Summary

Both VAEs and Diffusion Models are **principled probabilistic generative models**
rooted in the same variational lower bound:

$$\log p(\boldsymbol{x}) \geq \text{ELBO}(\boldsymbol{\phi},\boldsymbol{\theta})
= \underbrace{\mathbb{E}[\log p_{\boldsymbol{\theta}}(\boldsymbol{x}|\boldsymbol{h})]}_\text{reconstruction}
- \underbrace{D_{\mathrm{KL}}(q_{\boldsymbol{\phi}}\,\|\,p)}_\text{regularisation}$$

Their differences stem from choices made when applying this framework:

| Design choice | VAE | DDPM |
|:-------------|:----|:-----|
| How many latent levels? | 1 | $T = 300$–$1000$ |
| Is the encoder learned? | Yes | No (fixed schedule) |
| What is the bottleneck? | Low-dim $\boldsymbol{h}$ | None (full-dim $\boldsymbol{x}_t$) |
| What does the network predict? | $\boldsymbol{x}$ from $\boldsymbol{h}$ | $\boldsymbol{\epsilon}$ from $\boldsymbol{x}_t,t$ |

The practical consequence is that VAEs trade **sample quality for latent interpretability**,
while diffusion models trade **sampling speed for generation quality**.
**Latent Diffusion Models** (Stable Diffusion) elegantly sidestep this trade-off
by running a DDPM inside a VAE's latent space.
